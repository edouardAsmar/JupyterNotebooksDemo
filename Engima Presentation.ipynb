## Your boss now asks:
* What regions and cities does traffic come from?
* How many new users each day?
* What devices and OSs are accessing the website?

## Let's walk through how to protype answers to these questions

### First load in the relevant data

````python
rawdata = []
with open ("/Users/edouardasmar/regionanddeviceservicebox.csv", "rb") as csvfile:
    reader = csv.DictReader(csvfile, dialect="excel")
    for row in reader:
        rawdata.append(row)
````

````python
df2 = pd.DataFrame(rawdata)
````

````python
df2
````

````python
df2[['bounces', 'entrances', 'newUsers', 'sessions', 'users']] = df2[['bounces', 'entrances', 'newUsers', 'sessions', 'users']].astype(int)
````

````python
df2.dtypes
````

### Finding the top 5 cities with their associated regions:
* First select users, city and region fields from the table
* Second group by region, then city (order matters!)
* Use the sum function to add up relevant users
* Sort the values by the sums
* Take the bottom 5 (Sort is ascending)

````python
usersPerCity = df2[['users', 'city', 'region']].groupby(['region', 'city']).sum().sort_values(by='users')
usersPerCity.tail()
````

### Creating a pie chart with the information
* Simply call ```.plot.pie(subplots=True, figsize=)```
* Subplots is ```True``` is the default

````python
#Top 5 region and city combinations
usersPerCity.tail().plot.pie(subplots=True, figsize=(15, 8))
````

### A similar analysis of one variable

````python
usersPerRegion = df2[['users', 'region']].groupby(['region']).sum().sort_values(by='users')
usersPerRegion.tail().plot.pie(subplots=True,figsize=(15,8))
````

### In order to figure out new users each day:
* Create a typed date field in order to use default time-series behavior
* Create a new table using the same methods as before
* Use ```.plot()``` for the visualization with no arguments as time-series line graphs are the default

````python
df['date'] = pd.to_datetime(df['date'],infer_datetime_format=True)#format="%y-%m-%d")
df['date']
newUsersPerDay = df[['newUsers','date','users']].groupby('date', as_index=True).sum()
newUsersPerDay
````

````python
newUsersPerDay.plot()
````

### Answering the devices and OSs question
* Create a new table with the desired data
* ```.plot()```!

````python
devicesAndCategories = df2[['deviceCategory', 'operatingSystem', 'users']].groupby(['deviceCategory', 'operatingSystem']).sum()
devicesAndCategories.plot.pie(subplots=True, figsize=(15, 8), autopct='%.2f')
````

### With one variable

````python
df2[['deviceCategory','users']].groupby(['deviceCategory']).sum().plot.pie(subplots=True, figsize=(15,8), autopct='%.2f')
````

### Your boss: "That's too much data! Can I just look at desktop users?"
#### You: "Yes, of course"

### How?
* Select conditions: ```.isin()```
* Use ```autopct=``` to show percentages on the chart

````python
desktopBrowserAndOS = df2[df2['deviceCategory'].isin(['desktop'])][['browser', 'users', 'operatingSystem']]
desktopBrowserAndOS.groupby(['operatingSystem']).sum().plot.pie(subplots=True, figsize=(15,8), autopct='%.2f')
````

### Your boss: "What browsers are they using on desktops?"
#### You: "Got it!"

````python
desktopBrowserAndOS.groupby(['operatingSystem', 'browser']).sum().plot.pie(subplots=True, figsize=(15,8), autopct='%.2f')
````

### Your boss: "...just browsers?"

````python
desktopBrowserAndOS.groupby(['browser']).sum().head().plot.pie(subplots=True, figsize=(15,8), autopct='%.2f')
````
